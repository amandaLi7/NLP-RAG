# QA Extraction

```python
# link: https://python.langchain.com/docs/use_cases/chatbots/retrieval

from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage
```

```python
def qa_gen_answer(
    question: str,
    docs_processed: LangchainDocument,
    ) -> Tuple[str, List[LangchainDocument], List[str], List[str]]:

    relevant_docs = [doc.page_content for doc in docs_processed]
    relevant_doc_id = [doc.metadata['source'] for doc in docs_processed]
    relevant_doc_index = [doc.metadata['start_index'] for doc in docs_processed]

    # Build the final prompt
    context = "\nExtracted documents:\n"
    context += "".join([f"Document {str(relevant_docs)}:::\n"])

    SYSTEM_TEMPLATE = """
    Answer the user's questions based on the context from documents below. Please keep them to the point and concise, and only give the answer.
    For example, if the context is about the solar system and the question is "What is the largest planet?", the answer should be "Jupiter".
    If the context doesn't contain any relevant information to the question, don't make something up and just say "I don't know"
    ------------
    {context}
    ------------
    Keep answers short, long answers only when really needed.
    """

    question_answering_prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                SYSTEM_TEMPLATE,
            ),
            MessagesPlaceholder(variable_name="messages"),
        ]
    )

    document_chain = create_stuff_documents_chain(llm_answer_gen, question_answering_prompt)
    answer = document_chain.invoke(
        {
            "context": docs_processed,
            "messages": [
                HumanMessage(content=question)
            ],
        }
    )
    print("=> Generating answer...")
    return answer, relevant_docs, relevant_doc_id, relevant_doc_index
```

### Creating questions by combining two chunks at a time

```python
qa_df_list = []

# for i in range(len(docs_processed)-1,2):
for i in range(4,6):
  docs = docs_processed[i: i + 2]
  question_gen_chain = load_summarize_chain(llm=llm_question_gen, prompt=PROMPT_QUESTIONS)
  questions = question_gen_chain.invoke(docs)

  question_output = questions['output_text'].split('\n')

  for question in question_output:
    # print(f'question: {question}')

    if question != '':
        try:
            try:
                question = question.split('. ', 1)[1]
            except ValueError:
                pass  # If splitting fails, proceed with the original question

            print(question)
            answer, relevant_docs, relevant_doc_id, relevant_doc_index = qa_gen_answer(question, docs)
            # print('ansnwer unprocessed', answer)
            answer = answer.split(': ', 1)[1]
            print(f'Question: {question}')
            print(f'Answer: {answer}')
            qa_df_list.append({"Question": question, "Answer": answer,'DocumentId': relevant_doc_id, 'StartIndex': relevant_doc_index})

        except Exception as e:
            print(f"Error processing question: {question}, Error: {e}")
    else:
      print('No good question found')

    print(qa_df_list)
```

```python
# Convert to dataframe
qa_df = pd.DataFrame(qa_df_list, columns=["Question", "Answer", 'DocumentId', 'StartIndex'])
qa_df
```
